```python
# top 5 honey producing states of all time
# data.groupby('column').aggregate
df1=df.groupby('state').sum()['production'].reset_index()
```

```python
df1
```

```python
# sorting data
# data.sort_vaues(by=column,ascendin=true/false)----> syntax
df1=df1.sort_values(by='production',ascending=False )# descending order
df1.head()
```

```python
# this is to make your figure large or small
plt.figure(figsize=(20,10))
# helps to rotate values in x-axis, rotation=90--> degree of rotation
plt.xticks(rotation=45)
#sns.barplot(data,x,y)-----> syntax
sns.barplot(data=df1,x='state',y='production')
# to hide unwanted data
plt.show()
```

```python
# this is to make your figure large or small
plt.figure(figsize=(20,10))
# helps to rotate values in x-axis, rotation=90--> degree of rotation
plt.xticks(rotation=45)
#sns.barplot(data,x,y)-----> syntax
sns.barplot(data=df1.head(),x='state',y='production')
# to hide unwanted data
plt.show()
```
